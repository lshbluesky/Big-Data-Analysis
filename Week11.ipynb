### 5.1장 - 예제 5-2 (titanic 데이터의 기본 정보 출력)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')

print(df['survived'].value_counts())
print(df['alive'].value_counts())
plt.pie(df['alive'].value_counts(), labels = ['no', 'yes'], explode = (0, 0.1), startangle = 90, shadow = True, autopct = '%.1f')
plt.show()
```

```python
print(df['sex'].value_counts())
plt.pie(df['sex'].value_counts(), labels = ['male', 'female'], explode = (0, 0.1), startangle = 90, shadow = True, autopct = '%.1f')
plt.show()
```

```python
print(df['who'].value_counts())
plt.pie(df['who'].value_counts(), labels = ['man', 'woman', 'child'], startangle = 90, shadow = True, autopct = '%.1f')
plt.show()
```

```python
print(df['embark_town'].value_counts())
plt.pie(df['embark_town'].value_counts(), labels = ['Southampton', 'Cherbourg', 'Queenstown'], startangle = 90, shadow = True, autopct = '%.1f')
plt.show()
```

```python
print(df['class'].value_counts())
plt.pie(df['class'].value_counts(), labels = ['Third', 'First', 'Second'], startangle = 90, autopct = '%.1f')
plt.show()
```

```python
print(df['alone'].value_counts())
plt.pie(df['alone'].value_counts(), labels = ['True', 'False'], startangle = 90, shadow = True, autopct = '%.1f')
plt.show()
```

### 5.1장 - 예제 5-3 (성별 생존자 수와 생존율의 비교)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')

sns.countplot(data = df, x = df['sex'], hue = df['alive'])
plt.show()

print(pd.crosstab(df['sex'], df['alive'], margins = True))
```

### 5.1장 - 예제 5-4 (객실 등급별 생존자 수와 생존율의 비교)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')

sns.countplot(data = df, x = df['class'], hue = df['alive'])
plt.show()
```

```python
print(pd.crosstab(df['class'], df['alive'], margins = True))
```

```python
sns.countplot(data = df, x = df['class'], hue = df['sex'])
plt.show()
```

```python
print(pd.crosstab(df['alive'], [df['class'], df['sex']], margins = True))
```

```python
sns.catplot(x = 'class', y = 'survived', col = 'sex', data = df, kind = 'bar')
plt.show()
```

### 5.1장 - 예제 5-5 (연령별 생존자 수와 생존율의 비교)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')

# 성별에 따른 나이 분포
g = sns.FacetGrid(df, col = 'sex')
g.map(plt.hist, 'age')
plt.show()
```

```python
# 성별과 생존 여부에 따른 나이 분포
g = sns.FacetGrid(df, col = 'sex', row = 'alive')
g.map(plt.hist, 'age')
plt.show()
```

```python
# 객실 등급에 따른 나이 분포
g = sns.FacetGrid(df, col = 'class')
g.map(plt.hist, 'age')
plt.show()
```

```python
# 객실 등급과 생존 여부에 따른 나이 분포
g = sns.FacetGrid(df, col = 'class', row = 'alive')
g.map(plt.hist, 'age')
plt.show()
```

### 5.1장 - 예제 5-6 (탑승 도시별 생존자 수와 생존율의 비교)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')

# 도시별 탑승 인원
sns.countplot(df, x = 'embark_town')
plt.show()
```

```python
# 탑승 도시별 성별 구분
sns.countplot(df, x = 'embark_town', hue = 'sex')
plt.show()
```

```python
# 탑승 도시별 생존 여부 구분
sns.countplot(df, x = 'embark_town', hue = 'alive')
plt.show()
```

```python
# 탑승 도시별 등급 구분
sns.countplot(df, x = 'embark_town', hue = 'class')
plt.show()
```

```python
# 탑승 도시와 성별에 따른 생존 여부의 사례수
print(pd.crosstab(df['alive'], [df['embark_town'], df['sex']], margins = True))
```

```python
# 탑승 도시와 등급에 따른 생존 여부의 사례수
print(pd.crosstab(df['alive'], [df['embark_town'], df['class']], margins = True))
```
