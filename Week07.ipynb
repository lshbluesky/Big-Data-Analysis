### 2.3.1장 - 히스토그램

### 2.3.1장 - 막대그래프

### 2.3.1장 - 파이 차트

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'

ratio = [46.0, 30.0, 13.6, 8.7, 1.7]
labels = ['고향 및 부모친척방문', '집에서 휴식', '국내여행', '해외여행', '기타']
plt.pie(ratio, labels = labels, autopct = '%.1f%%') # 뒤에 퍼센트 기호를 같이 표시
plt.title('추석 연휴계획 조사 결과')
plt.show()
```

### 2.3.1장 - 선 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.DataFrame(pd.read_excel("BigData_Sources/휘발유_우유가격변화.xlsx"))
plt.rcParams['font.family'] = 'Malgun Gothic'

x = data.year
y1 = data.gasoline
y2 = data.milk

plt.plot(x, y1, linestyle = 'solid', label = 'gasoline')
plt.plot(x, y2, linestyle = 'dashed', label = 'milk')

plt.legend(loc = 'best', ncol = 2)
plt.title('Change of price')
plt.show()
```

### 2.3.2장 - 중학교 남학생들의 몸무게에 대한 대푯값 출력 예제

```python
import numpy as np
import pandas as pd
from scipy import stats
x = pd.read_excel("BigData_Sources/중학생_남자_몸무게.xlsx")

print('mean :', np.average(x))
print('median :', np.median(x))
print('mode :', stats.mode(x))
```

### 2.3.2장 - 산포도 출력 예제

```python
import numpy as np
import pandas as pd
data = pd.DataFrame(pd.read_excel("BigData_Sources/고등학생_samples.xlsx"))

print('키의 범위 :', np.max(data.height)-np.min(data.height))
print('몸무게의 범위  :', np.max(data.weight)-np.min(data.weight), '\n')

print('키의 분산 :', np.var(data.height))
print('몸무게의 분산 :', np.var(data.weight), '\n')

print('키의 표준편차 :', np.std(data.height))
print('몸무게의 표준편차 :', np.std(data.weight), '\n')
```
